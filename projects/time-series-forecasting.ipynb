```python
result = seasonal_decompose(cars.set_index('date')['value'], model='additive')

fig, ax = plt.subplots(4, 1, figsize = (12, 8))
sns.lineplot(result.observed, ax = ax[0]).set(xlabel = '', ylabel = 'observed', xticklabels=[]); ax[0].tick_params(bottom=False);
sns.lineplot(result.trend, ax = ax[1]).set(xlabel = '', xticklabels=[]); ax[1].tick_params(bottom=False);
sns.lineplot(result.seasonal, ax = ax[2]).set(xlabel = '', xticklabels=[]); ax[2].tick_params(bottom=False);
sns.lineplot(result.resid, ax = ax[3]);
```

Seasonal decomposition is useful in visualizing the underlying trend and seasonality of the time series. The vehicle sales seem to have a fluctuating trend where it increases and decreases periodically. It also shows a clear seasonality. The residuals show periods of high variability in the late 1980s and early 2020s.

## Feature engineering

```python
df = cars.merge(gas, how = 'inner', on = 'date', suffixes = ('_cars', '_gas'))
df = df[['date', 'value_cars', 'value_gas']]
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

df['gas_rolling_3d_mean'] = df['value_gas'].rolling(window = 3, min_periods = 0).mean().shift(1).fillna(df['value_gas'])
df['gas_rolling_7d_mean'] = df['value_gas'].rolling(window = 7, min_periods = 0).mean().shift(1).fillna(df['value_gas'])
df['gas_rolling_30d_mean'] = df['value_gas'].rolling(window = 30, min_periods = 0).mean().shift(1).fillna(df['value_gas'])
#df['gas_rolling_365d_mean'] = df['value_gas'].rolling(window = 365, min_periods = 0).mean().shift(1).fillna(df['value_gas'])

df['gas_rolling_3d_std'] = df['value_gas'].rolling(window = 3, min_periods = 0).std().shift(1).fillna(0)
df['gas_rolling_7d_std'] = df['value_gas'].rolling(window = 7, min_periods = 0).std().shift(1).fillna(0)
df['gas_rolling_30d_std'] = df['value_gas'].rolling(window = 30, min_periods = 0).std().shift(1).fillna(0)
#df['gas_rolling_365d_std'] = df['value_gas'].rolling(window = 365, min_periods = 0).std().shift(1).fillna(0)

df['gas_lagged_1d'] = df['value_gas'].shift(1).fillna(df['value_gas'])
df['gas_lagged_7d'] = df['value_gas'].shift(7).fillna(df['value_gas'])
df['gas_lagged_30d'] = df['value_gas'].shift(30).fillna(df['value_gas'])
#df['gas_lagged_365d'] = df['value_gas'].shift(365).fillna(df['value_gas'])

df = df.set_index('date')
df
```

## Train test split

We are going to forecast the total vehicle sales in US from 2021. So let's split the dataset to train and test set.

```python
train = df.loc[df.index < '2023-01-01']
test = df.loc[df.index >= '2023-01-01']

exog_features = train.columns[1:].tolist()

X_train, X_test, y_train, y_test = train[exog_features], test[exog_features], train['value_cars'], test['value_cars']
```

## XGBoost

```python
import xgboost
```

```python
xgb = xgboost.XGBRegressor(random_state = 5)
xgb = xgb.fit(X_train, y_train)
forecast = xgb.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.lineplot(train['value_cars']).set(ylabel = 'Vehicle sales');
sns.lineplot(test['value_cars'], label = 'Actual');
sns.lineplot(pd.DataFrame(forecast).set_index(test.index)[0], label = 'Forecast');
```

```python
print("RMSE of XGB:", np.sqrt(mean_squared_error(test['value_cars'], forecast)))
print("MAE of XGB:", mean_absolute_error(test['value_cars'], forecast))
```

## LightGBM

```python
params = {'objective': 'regression', 'verbose': -1}
dtrain = lgb.Dataset(X_train, label = y_train)
model_lgb = lgb.train(params, train_set = dtrain)
forecast = model_lgb.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.lineplot(train['value_cars']).set(ylabel = 'Vehicle sales');
sns.lineplot(test['value_cars'], label = 'Actual');
sns.lineplot(pd.DataFrame(forecast).set_index(test.index)[0], label = 'Forecast');
```

```python
print("RMSE of LightGBM:", np.sqrt(mean_squared_error(test['value_cars'], forecast)))
print("MAE of LightGBM:", mean_absolute_error(test['value_cars'], forecast))
```

## LSTM

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train_nn = scaler.fit_transform(X_train)
X_test_nn = scaler.fit_transform(X_test)
```

```python
X_train_nn = X_train_nn.reshape((X_train_nn.shape[0], X_train_nn.shape[1], 1))
X_test_nn = X_test_nn.reshape((X_test_nn.shape[0], X_test_nn.shape[1], 1))
```

```python
series_size =  X_train_nn.shape[1] # 12
n_features =  X_train_nn.shape[2] # 1

epochs = 100
batch = 32
lr = 0.01

lstm_model = Sequential()
lstm_model.add(L.LSTM(12, input_shape=(series_size, n_features), return_sequences=True))
#lstm_model.add(L.LSTM(6, activation='relu', return_sequences=True))
lstm_model.add(L.LSTM(1, activation='relu'))
lstm_model.add(L.Dense(12, kernel_initializer='glorot_normal', activation='relu'))
#lstm_model.add(L.Dense(12, kernel_initializer='glorot_normal', activation='relu'))
lstm_model.add(L.Dense(1))
lstm_model.summary()

adam = optimizers.Adam(lr)
lstm_model.compile(loss='mse', optimizer=adam)
```

```python
callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

lstm_history = lstm_model.fit(X_train_nn, y_train, 
                              validation_data=(X_test_nn, y_test), 
                              batch_size=batch, 
                              epochs=epochs, 
                              callbacks=[callback], 
                              verbose=2)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.lineplot(lstm_history.history['loss'], label='Train loss');
sns.lineplot(lstm_history.history['val_loss'], label='Validation loss');
```

```python
forecast = lstm_model.predict(X_test_nn)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.lineplot(train['value_cars']).set(ylabel = 'Vehicle sales');
sns.lineplot(test['value_cars'], label = 'Actual');
sns.lineplot(pd.DataFrame(forecast).set_index(test.index)[0], label = 'Forecast');
```

```python
print("RMSE of LSTM:", np.sqrt(mean_squared_error(test['value_cars'], forecast)))
print("MAE of LSTM:", mean_absolute_error(test['value_cars'], forecast))
```

## Prophet

```python
exog_feat = ['value_gas']
```

```python
model_fbp = Prophet()
for feature in exog_feat:
    model_fbp.add_regressor(feature)

model_fbp.fit(train.reset_index()[["date", "value_cars"] + exog_feat].rename(columns={"date": "ds", "value_cars": "y"}))

forecast = model_fbp.predict(test.reset_index()[["date", "value_cars"] + exog_feat].rename(columns={"date": "ds"}))
```

```python
model_fbp.plot_components(forecast)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.lineplot(train['value_cars']).set(ylabel = 'Vehicle sales');
sns.lineplot(test['value_cars'], label = 'Actual');
sns.lineplot(pd.DataFrame(forecast.yhat.values).set_index(test.index)[0], label = 'Forecast');
sns.lineplot(pd.DataFrame(forecast.yhat_upper.values).set_index(test.index)[0], label = 'Upper 80% interval');
sns.lineplot(pd.DataFrame(forecast.yhat_lower.values).set_index(test.index)[0], label = 'Lower 80% interval');
ax.fill_between(test.index, forecast.yhat_lower.values, forecast.yhat_upper.values, color = 'gray', alpha = 0.3);
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.lineplot(test['value_cars'], label = 'Actual').set(ylabel = 'Vehicle sales');
sns.lineplot(pd.DataFrame(forecast.yhat.values).set_index(test.index)[0], label = 'Forecast');
```

```python
print("RMSE of Prophet:", np.sqrt(mean_squared_error(test['value_cars'], forecast.yhat.values)))
print("MAE of Prophet:", mean_absolute_error(test['value_cars'], forecast.yhat.values))
```
